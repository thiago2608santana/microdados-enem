```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('ITENS_PROVA_2019.csv',sep=';',encoding='latin-1')
```

```python
dados.head()
```

```python
plt.figure(figsize=(14,6))
grafico_linguagens_codigos = sns.countplot('CO_HABILIDADE', data=dados[dados['SG_AREA'] == 'LC'])
plt.xlabel('Código da habilidade', fontweight='bold', fontsize=14)
plt.ylabel('Quantidade', fontweight='bold', fontsize=14)
plt.title('Habilidades cobradas pela área de Linguagens, Códigos - Enem 2019', fontweight='bold', fontsize=16)
plt.savefig('grafico_linguagens_codigos.png', dpi=400)
#H6 - Utilizar os conhecimentos da LEM e de seus mecanismos como meio de ampliar as
#possibilidades de acesso a informações, tecnologias e culturas

#H16 - Relacionar informações sobre concepções artísticas e procedimentos de construção
#do texto literário
```
